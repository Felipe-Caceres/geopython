**Comenzaremos creando algunos gráficos de "coordenada" vs "variable de interés". En este caso, Este - Cu, Norte -Cu, Elevación - Cu** 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
carpeta_origen = "C:/Users/Heber/Desktop/curso_geopython/dataset_curso/"
archivo_1 = "muestras_cu.csv"

df = pd.read_csv(carpeta_origen + archivo_1, sep=";")

# Inicialmente nos interesa conocer los min() y max() de los ejes (x,y,z)

#Nos interesa conocer el comportamiento de la variable "Cu" a distintas distancias en los ejes antes mencionados.
```

**En nuestro caso, utilizando unicamente las librerias antes importadas, es que requerimos de crear algunos df:**

```python
# Matriz Este - Cobre

df_new_x = pd.DataFrame({'From': list(range(0,351,50)), 'To': list(range(50,401,50)), 'Cu': np.nan})

# Debemos calcular la ley media de cobre en cada una de las ventanas creadas (50 metros):

a = (df[(df['Este [X]']>= 0) & (df['Este [X]']<50)]).mean()
ax = a['Ley de Cobre [%]'] 
b = (df[(df['Este [X]']>= 50) & (df['Este [X]']<100)]).mean()
bx= b['Ley de Cobre [%]'] 
c = (df[(df['Este [X]']>= 100) & (df['Este [X]']<150)]).mean()
cx = c['Ley de Cobre [%]'] 
d = (df[(df['Este [X]']>= 150) & (df['Este [X]']<200)]).mean()
dx = d['Ley de Cobre [%]'] 
e = (df[(df['Este [X]']>= 200) & (df['Este [X]']<250)]).mean()
ex = e['Ley de Cobre [%]'] 
f = (df[(df['Este [X]']>= 250) & (df['Este [X]']<300)]).mean()
fx= f['Ley de Cobre [%]'] 
g = (df[(df['Este [X]']>= 300) & (df['Este [X]']<350)]).mean()
gx=g['Ley de Cobre [%]'] 
h = (df[(df['Este [X]']>= 350) & (df['Este [X]']<400)]).mean()
hx=h['Ley de Cobre [%]'] 
# los nx son las leyes medias de Cu para cada ventana del eje "X", recordar que la longitud por ventana es de 50 m.
# ahora pasaremos dichas leyes de Cu al df:
df_new_x['Cu'] = [ax,bx,cx,dx,ex,fx,gx,hx] #pase de las leyes de Cu al df original.
# Ahora ya podemos gráficar, sin embargo, creare una nueva columna con el promedio de "from -to":
df_new_x['promxy'] = df_new_x['To'] - 25 # promedio from -to


# Matriz Norte - Cobre [ventana tambien de 50 m]

df_new_y = pd.DataFrame({'From':list(range(0,551,50)), 'To':list(range(50,601,50)), 'Cu [%]': np.nan})

a2 = (df[(df["Norte [Y]"]>=0) & (df["Norte [Y]"] <50)]).mean()
ay = a2['Ley de Cobre [%]']
b2 = df[(df["Norte [Y]"]>=50) & (df["Norte [Y]"] <100)].mean()
by = b2['Ley de Cobre [%]']
c2 = df[(df["Norte [Y]"]>=100) & (df["Norte [Y]"] <150)].mean()
cy = c2['Ley de Cobre [%]']
d2 = df[(df["Norte [Y]"]>=150) & (df["Norte [Y]"] <200)].mean()
dy = d2['Ley de Cobre [%]']
e2 = df[(df["Norte [Y]"]>=200) & (df["Norte [Y]"] <250)].mean()
ey = e2['Ley de Cobre [%]']
f2 = df[(df["Norte [Y]"]>=250) & (df["Norte [Y]"] <300)].mean()
fy = f2['Ley de Cobre [%]']
g2 = df[(df["Norte [Y]"]>=300) & (df["Norte [Y]"] <350)].mean()
gy = g2['Ley de Cobre [%]']
h2 = df[(df["Norte [Y]"]>=350) & (df["Norte [Y]"] <400)].mean()
hy = h2['Ley de Cobre [%]']
y2 = df[(df["Norte [Y]"]>=400) & (df["Norte [Y]"] <450)].mean()
yy = y2['Ley de Cobre [%]']
j2 = df[(df["Norte [Y]"]>=450) & (df["Norte [Y]"] <500)].mean()
jy = j2['Ley de Cobre [%]']
k2 = df[(df["Norte [Y]"]>=500) & (df["Norte [Y]"] <550)].mean()
ky = k2['Ley de Cobre [%]']
l2 = df[(df["Norte [Y]"]>=550) & (df["Norte [Y]"] <600)].mean()
ly = l2['Ley de Cobre [%]']


df_new_y['Cu [%]'] = [ay,by,cy,dy,ey,fy,gy,hy,yy,jy,ky,ly] #la cantidad de datos en Norte - Sur es mayor que en el E - W

df_new_y['promxy'] = df_new_y['To'] - 25

# Matriz Cota - Cobre [ventana de 20 m dado que la distancia va entre 0 y 140 m]

df_new_z = pd.DataFrame({'From':list(range(0,121,20)), 'To':list(range(20,141,20)), 'Cu [%]': np.nan})

a3 = (df[(df["Cota [Z]"]>=0) & (df["Cota [Z]"] <20)]).mean()
az = a3['Ley de Cobre [%]']
b3 = (df[(df["Cota [Z]"]>=20) & (df["Cota [Z]"] <40)]).mean()
bz = b3['Ley de Cobre [%]']
c3 = (df[(df["Cota [Z]"]>=40) & (df["Cota [Z]"] <60)]).mean()
cz = c3['Ley de Cobre [%]']
d3 = (df[(df["Cota [Z]"]>=60) & (df["Cota [Z]"] <80)]).mean()
dz = d3['Ley de Cobre [%]']
e3 = (df[(df["Cota [Z]"]>=80) & (df["Cota [Z]"] <100)]).mean()
ez = e3['Ley de Cobre [%]']
f3 = (df[(df["Cota [Z]"]>=100) & (df["Cota [Z]"] <120)]).mean()
fz = f3['Ley de Cobre [%]']
g3 = (df[(df["Cota [Z]"]>=120) & (df["Cota [Z]"] <140)]).mean()
gz = g3['Ley de Cobre [%]']

df_new_z['Cu [%]'] = [az,bz,cz,dz,ez,fz,gz]

df_new_z['promxy'] = df_new_z['To'] - 10
```

```python
# Visualizaremos 3 gráficos en 1 figura:

figura = plt.figure(figsize=(20,5))

figura.suptitle("DERIVAS LEY DE COBRE [%] DATASET N° 1")

plot1 = figura.add_subplot(131)
plt.plot(df_new_x['promxy'], df_new_x['Cu'])
plt.xlabel("Este [X]")
plt.ylabel("Ley de Cobre [%]")
plt.title("Este - Oeste ")
plt.grid(True)

plot2 = figura.add_subplot(132)
plt.plot(df_new_y['promxy'], df_new_y['Cu [%]'])
plt.xlabel("Norte [Y]")
plt.ylabel("Ley de Cobre [%]")
plt.title("Norte - Sur ")
plt.grid(True)

plot3 = figura.add_subplot(133)
plt.plot(df_new_z['promxy'], df_new_z['Cu [%]'])
plt.xlabel("Cota [Z]")
plt.ylabel("Ley de Cobre [%]")
plt.title("Elevación ")
plt.grid(True)

plt.show()
```
